```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/CV_nangcao/train/Retinanet/result.csv')
print(df.to_string()) 
```

```python
epoch = df['epoch']
figure, axis = plt.subplots(3, 4,figsize=(25,15))
#plot Metric
mAP_05= df['mAP']
axis[0, 0].plot(epoch, mAP_05)
axis[0, 0].set_title("mAP@.5")

#plot Loss

val_total_loss=df['val_loss']
train_total_loss=df['loss']
axis[0, 1].plot(epoch, train_total_loss,'r',label='train')
axis[0, 1].plot(epoch, val_total_loss,'g',label='val')
axis[0, 1].legend(loc='upper right')
axis[0, 1].set_title("Total loss")

val_reg_loss=df['val_reg_loss']
train_reg_loss=df['regression_loss']
axis[0, 2].plot(epoch, train_reg_loss,'r',label='train')
axis[0, 2].plot(epoch, val_reg_loss,'g',label='val')
axis[0, 2].legend(loc='upper right')
axis[0, 2].set_title("Regression loss")

val_classification_loss=df['val_classification_loss']
train_classification_loss=df['classification_loss']
axis[0, 3].plot(epoch, train_classification_loss,'r',label='train')
axis[0, 3].plot(epoch, val_classification_loss,'g',label='val')
axis[0, 3].legend(loc='upper right')
axis[0, 3].set_title("Classification loss")


plt.show()
```
