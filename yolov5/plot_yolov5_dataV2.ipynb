```python
import pandas as pd
import matplotlib.pyplot as plt
import time
from IPython.display import clear_output
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
path='/content/drive/MyDrive/CV_nangcao/train/yolov5/yolov5_result_dataV2/exp3/results.csv'
```

```python
while True:
  log=pd.read_csv(path)
  epoch=log['               epoch']
  train_box_loss=log['      train/box_loss']
  train_cls_loss=log['      train/cls_loss']
  train_obj_loss=log['      train/obj_loss']
  train_total_loss=train_box_loss+train_cls_loss+train_obj_loss
  val_box_loss=log['        val/box_loss']
  val_cls_loss=log['        val/cls_loss']
  val_obj_loss=log['        val/obj_loss']
  val_total_loss=val_box_loss+val_cls_loss+val_obj_loss
  mAP_05=log['     metrics/mAP_0.5']
  mAP_05_095=log['metrics/mAP_0.5:0.95']
  precision=log['   metrics/precision']
  recall=log['      metrics/recall']
  lr0=log['               x/lr0']
  lr1=log['               x/lr1']
  lr2=log['               x/lr2']
  figure, axis = plt.subplots(3, 4,figsize=(25,15))
  #plot Metric
  axis[0, 0].plot(epoch, mAP_05)
  axis[0, 0].set_title("mAP@.5")
  axis[0, 1].plot(epoch, mAP_05_095)
  axis[0, 1].set_title("mAP@.5:.95")
  axis[0, 2].plot(epoch, precision)
  axis[0, 2].set_title("Precision")
  axis[0, 3].plot(epoch, recall)
  axis[0, 3].set_title("Recall")

  #plot Loss
  axis[1, 0].plot(epoch, train_total_loss,'r',label='train')
  axis[1, 0].plot(epoch, val_total_loss,'g',label='val')
  axis[1, 0].legend(loc='upper right')
  axis[1, 0].set_title("Total loss")
  axis[1, 1].plot(epoch, train_box_loss,'r',label='train')
  axis[1, 1].plot(epoch, val_box_loss,'g',label='val')
  axis[1, 1].legend(loc='upper right')
  axis[1, 1].set_title("Box loss")
  axis[1, 2].plot(epoch, train_cls_loss,'r',label='train')
  axis[1, 2].plot(epoch, val_cls_loss,'g',label='val')
  axis[1, 2].legend(loc='upper right')
  axis[1, 2].set_title("Classification loss")
  axis[1, 3].plot(epoch, train_obj_loss,'r',label='train')
  axis[1, 3].plot(epoch, val_obj_loss,'g',label='val')
  axis[1, 3].legend(loc='upper right')
  axis[1, 3].set_title("Object loss")

  #plot Learning rate
  axis[2, 0].plot(epoch, lr0)
  axis[2, 0].set_title("x/lr0")
  axis[2, 1].plot(epoch, lr1)
  axis[2, 1].set_title("x/lr1")
  axis[2, 2].plot(epoch, lr2)
  axis[2, 2].set_title("x/lr2")
  axis[2, 3].set_visible(False)
  plt.show()
  time.sleep(60)
  clear_output(wait=True)
```
